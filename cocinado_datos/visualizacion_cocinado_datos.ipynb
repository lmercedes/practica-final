```python
from azureml import Workspace
import numpy as np

ws = Workspace()
ds_2 = ws.datasets['kiva_loans.csv']
kiva_loans = ds_2.to_dataframe()
```

```python
kiva_loans.head(10)
```

```python
prestamos_pais = kiva_loans.groupby(['country']).size().to_frame(name = 'cantidad_prestamos').reset_index()
prestamos_pais.sort_values(['cantidad_prestamos'], ascending=False)
prestamos_pais = prestamos_pais[prestamos_pais['cantidad_prestamos'] > 50 ].sort_values(['cantidad_prestamos'], ascending=False)
prestamos_pais 
```

```python
kiva_loans['disbursed_year'] = kiva_loans['disbursed_time'].str.split('-',expand=True)[0]
kiva_loans['disbursed_date'] = kiva_loans['disbursed_time'].str.split(' ',expand=True)[0]
```

```python
prestamos_pais_fecha = kiva_loans.groupby(['country','country_code','disbursed_year']).size().to_frame(name = 'cantidad_prestamos').reset_index()
prestamos_pais_fecha = prestamos_pais_fecha.sort_values(['disbursed_year'], ascending=True)
prestamos_pais_fecha = prestamos_pais_fecha[prestamos_pais_fecha['cantidad_prestamos'] > 10]
prestamos_pais_fecha = prestamos_pais_fecha[prestamos_pais_fecha.columns[0:5]]
prestamos_pais_fecha.head(10)
```

```python
prestamos_pais_top_10 = prestamos_pais_fecha.loc[prestamos_pais_fecha['country'].isin(['Philippines', 'Kenya','Cambodia', 'El Salvador', 'Peru'])]
prestamos_pais_top_10 = prestamos_pais_top_10[prestamos_pais_top_10.columns[0:5]]
prestamos_pais_top_10.head(10)
```

```python
#JSON Paises
countries = prestamos_pais_fecha[['country_code', 'country']]
countries = countries.sort_values(['country_code'], ascending=True).drop_duplicates()
countries.to_json(orient='records')    
```

```python
# JSON Prestamos por Pais
prestamos_pais_json =  prestamos_pais.to_json(orient='records')
prestamos_pais_json
```

```python
#JSON Prestamos Top 10
prestamos_pais_top_10_json =  prestamos_pais_top_10.to_json(orient='records')
prestamos_pais_top_10_json
```

```python
#JSON Prestamo Fecha
prestamos_pais_fecha_json =  prestamos_pais_fecha.to_json(orient='records')
prestamos_pais_fecha_json
```
